```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Order To Collection.xlsx')
```

```python
df = pd.DataFrame(data)
```

```python
df1 = pd.DataFrame(df[['Unique Key','Customer Name','Company Name','City Name','Total Price']])
```

```python
df1
```

```python
df1['Unique Key'].fillna('OTC000', inplace = True)
```

```python
df1['Unique Key'].isna()
```

```python
df1['Customer Name'].fillna('Kumar Narayan',inplace = True)
```

```python
df1['Company Name'].fillna('KK Group',inplace = True)
```

```python
df1['City Name'].fillna('Deoria',inplace = True)
```

```python
mn = df1['Total Price'].mean()
```

```python
mn
```

```python
df1['Total Price'].fillna(df1['Total Price'].mean(),inplace = True)
```

```python
df1
```

```python
x = df1['City Name']
y = df1['Total Price']

x1 = x[::200]
y1 = y[::200]

plt.barh(x1,y1)
plt.xlabel('Sales')
plt.ylabel('City Name')
plt.title('Sales Graph')
plt.bar_label(plt.barh(x1,y1))
```

```python
x = df1['City Name']
y = df1['Total Price']

x1 = x[::200]
y1 = y[::200]

plt.pie(y1,labels = x1 , autopct ="%0.2f%%")
plt.title('Sales Graph')
```

```python
plt.bar(df1['City Name'],df1['Total Price'])
plt.xlabel('City Name')
plt.ylabel('Sales')
plt.title('Sales Graph')
x = df1['City Name']
y = df1['Total Price']
for i in range(len(y)):
    plt.text(x[i],y[i],f'{y[i]}')
```
